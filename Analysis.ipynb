```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load de data from xlsx
path = "D:\\Invesitgacion\\Datathon\\Data Files\\train_data.xlsx"
df = pd.read_excel(path)
df
```

```python
print(f'Len of the data: {len(df)}')
print(f'Columns of the data: {df.columns}')

## Check for missing values
missing_values = df.isnull().sum()
print(f'Missing values: {missing_values}')

## Check for Nan values
nan_values = df.isna().sum()
print(f'Nan values: {nan_values}')

## Check unique values of cluster_nl
unique_values = df['cluster_nl'].unique()
print(f'Number of unique values in cluster_nl: {len(unique_values)}') 
```

```python
## Relationship between all numerical variables
## Take only the numerical variables
numerical_df = df.select_dtypes(include=[np.number])
## Correlation matrix
correlation_matrix = numerical_df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(numerical_df, alpha=0.2, figsize=(40,40), diagonal='kde')
plt.show()
```
