```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load de data from xlsx
path = "D:\\Invesitgacion\\Datathon\\Data Files\\V2\\train_data.csv"
df = pd.read_csv(path)
df
```

```python
print(f'Len of the data: {len(df)}')
print(f'Columns of the data: {df.columns}')

## Check for missing values
missing_values = df.isnull().sum()
print(f'Missing values: {missing_values}')

## Check for Nan values
nan_values = df.isna().sum()
print(f'Nan values: {nan_values}')

## Check unique values of cluster_nl
unique_values = df['cluster_nl'].unique()
print(f'Number of unique values in cluster_nl: {len(unique_values)}') 
```

```python
## Relationship between all numerical variables
## Take only the numerical variables
numerical_df = df.select_dtypes(include=[np.number])
## Correlation matrix
correlation_matrix = numerical_df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
## Corr values shorter than 1 and sort them

correlation_values = correlation_matrix.unstack().sort_values(ascending=False)
correlation_values = correlation_values[correlation_values != 1]
print(correlation_values)
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(numerical_df, alpha=0.2, figsize=(12,12), diagonal='kde')
plt.show()
```

```python
## Plot each numerical variable with the target variable
for column in numerical_df.columns:
    plt.scatter(df[column], df['target'])
    plt.xlabel(column)
    plt.ylabel('target')
    plt.show()
    
```

```python
## Take only the data with a date between 2019 and 2021 of launch_date
df['launch_date'] = pd.to_datetime(df['launch_date'])
df_covid = df[df['launch_date'].dt.year.between(2019, 2021)]
# Take only the data with a date without 2019 and 2021 of launch_date
df_no_covid = df[~df['launch_date'].dt.year.between(2019, 2021)]
df_covid
```

```python
## Analyze the target variable in df_covid and df_no_covid

plt.hist(df_covid['target'], bins=300)
plt.title('Target variable in df_covid')
plt.xlim(1, 3)
plt.show()

plt.hist(df_no_covid['target'], bins=300)
plt.title('Target variable in df_no_covid')
plt.xlim(1, 3)
plt.show()
```

```python
## Plot the che_perc_gdp mean per year
df['launch_date'] = pd.to_datetime(df['launch_date'])
df['year'] = df['launch_date'].dt.year
df.groupby('year')['che_perc_gdp'].mean().plot()
plt.title('che_perc_gdp mean per year')
plt.show()

## Plot the che_perc_gdp mean per month and year with a large visualization
df['month'] = df['launch_date'].dt.month
df.groupby(['year', 'month'])['che_perc_gdp'].mean().plot()
plt.title('che_perc_gdp mean per month and year')
plt.show()


```

```python
# Suponiendo que tu DataFrame se llama 'df' y ya tiene 'year' y 'month'
df['month'] = df['launch_date'].dt.month
df['year'] = df['launch_date'].dt.year

# Calcular la media agrupada por aÃ±o y mes
mean_values = df.groupby(['year', 'month'])['che_perc_gdp'].mean().reset_index()

# Crear una columna combinada para etiquetas del eje X
mean_values['year_month'] = mean_values['year'].astype(str) + '-' + mean_values['month'].astype(str)

# Crear el grÃ¡fico
plt.figure(figsize=(14, 7))
plt.plot(mean_values['year_month'], mean_values['che_perc_gdp'])

# Configurar etiquetas del eje X
plt.xticks(ticks=range(len(mean_values['year_month'])), labels=mean_values['year_month'], rotation=90, ha='center')

# TÃ­tulos y etiquetas
plt.title('che_perc_gdp mean per month and year')
plt.xlabel('Year-Month')
plt.ylabel('Mean che_perc_gdp')

# Mostrar el grÃ¡fico
plt.tight_layout()
plt.show()
```

```python
## Plot each year independently with all months 
for year in df['year'].unique():
    df_year = df[df['year'] == year]
    df_year.groupby('month')['che_perc_gdp'].mean().plot()
    # Add a red line for the mean of the year and the number over the line
    plt.axhline(df_year['che_perc_gdp'].mean(), color='red', linestyle='--')
    plt.text(11, df_year['che_perc_gdp'].mean(), f'{df_year["che_perc_gdp"].mean():.2f}')
    plt.title(f'che_perc_gdp mean per month in {year}')
    plt.show()
    
    ## Print the month with the lowest che_perc_gdp
    min_che_perc_gdp = df_year.groupby('month')['che_perc_gdp'].mean().idxmin()
    print(f'The month with the lowest che_perc_gdp in {year} is {min_che_perc_gdp}')
```
