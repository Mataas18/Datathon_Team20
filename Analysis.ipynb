```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load de data from xlsx
path = "Data Files\V2\\train_data.csv"
df = pd.read_csv(path)
df
```

```python
# Suponiendo que tu DataFrame se llama 'df' y ya tiene 'year' y 'month'
df['month'] = df['launch_date'].dt.month
df['year'] = df['launch_date'].dt.year

# Calcular la media agrupada por año y mes
mean_values = df.groupby(['year', 'month'])['che_perc_gdp'].mean().reset_index()

# Crear una columna combinada para etiquetas del eje X
mean_values['year_month'] = mean_values['year'].astype(str) + '-' + mean_values['month'].astype(str)

# Crear el gráfico
plt.figure(figsize=(14, 7))
plt.plot(mean_values['year_month'], mean_values['che_perc_gdp'])

# Configurar etiquetas del eje X
plt.xticks(ticks=range(len(mean_values['year_month'])), labels=mean_values['year_month'], rotation=90, ha='center')

# Títulos y etiquetas
plt.title('che_perc_gdp mean per month and year')
plt.xlabel('Year-Month')
plt.ylabel('Mean che_perc_gdp')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
## Plot each year independently with all months 
for year in df['year'].unique():
    df_year = df[df['year'] == year]
    df_year.groupby('month')['che_perc_gdp'].mean().plot()
    # Add a red line for the mean of the year and the number over the line
    plt.axhline(df_year['che_perc_gdp'].mean(), color='red', linestyle='--')
    plt.text(11, df_year['che_perc_gdp'].mean(), f'{df_year["che_perc_gdp"].mean():.2f}')
    plt.title(f'che_perc_gdp mean per month in {year}')
    plt.show()
    
    ## Print the month with the lowest che_perc_gdp
    min_che_perc_gdp = df_year.groupby('month')['che_perc_gdp'].mean().idxmin()
    print(f'The month with the lowest che_perc_gdp in {year} is {min_che_perc_gdp}')
```

```python
import pandas as pd

def process_dataframe_by_frequency(df, column, top_n=20):
    """
    Procesa un DataFrame agrupando las 20 muestras con más apariciones según una columna dada
    y suma el resto en una fila llamada 'ukn'.
    
    Args:
        df (pd.DataFrame): DataFrame de entrada.
        column (str): Nombre de la columna para calcular las apariciones.
    
    Returns:
        pd.DataFrame: Nuevo DataFrame con las 20 más frecuentes y una fila adicional 'ukn'.
    """
    # Contar las apariciones de cada valor en la columna
    value_counts = df[column].value_counts()
    
    # Seleccionar las 20 más frecuentes
    top_20 = value_counts.iloc[:top_n]
    
    # Calcular las apariciones del resto
    ukn_count = value_counts.iloc[top_n:].sum()
    
    # Crear un nuevo DataFrame con las 20 más frecuentes
    top_20_df = top_20.reset_index()
    top_20_df.columns = [column, 'Count']  # Renombrar las columnas
    
    # Agregar la fila 'ukn' al DataFrame
    ukn_row = pd.DataFrame({column: ['ukn'], 'Count': [ukn_count]})
    result_df = pd.concat([top_20_df, ukn_row], ignore_index=True)
    
    return result_df

```
